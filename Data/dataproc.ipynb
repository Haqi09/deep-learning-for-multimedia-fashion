```python
import pandas as pd
import os
import shutil

# Membaca file CSV
df = pd.read_csv('filtered_categories.csv')

# Definisi path
source_dir = '../Dataset/images'  # Folder sumber gambar
base_dest_dir = '../Dataset'  # Folder utama untuk kategori

# Membuat folder untuk setiap kategori
categories = ['Bags', 'Bottomwear', 'Dress', 'Headwear', 'Shoes', 'Topwear', 'Watches']

print("Membuat folder kategori...")
for category in categories:
    category_path = os.path.join(base_dest_dir, category)
    if not os.path.exists(category_path):
        os.makedirs(category_path)
        print(f"Membuat folder: {category}")

# Menghitung total file yang akan dipindahkan
total_files = len(df)
processed_files = 0

print("\nMemulai pemindahan file...")
# Memindahkan file ke folder yang sesuai
for index, row in df.iterrows():
    img_id = str(row['id'])
    category = row['subCategory']
    
    # Mencari file gambar (mencoba beberapa kemungkinan ekstensi)
    for ext in ['.jpg', '.jpeg', '.png']:
        source_file = os.path.join(source_dir, img_id + ext)
        if os.path.exists(source_file):
            dest_file = os.path.join(base_dest_dir, category, img_id + ext)
            try:
                shutil.copy2(source_file, dest_file)
                processed_files += 1
                # Update progress setiap 100 file
                if processed_files % 100 == 0:
                    progress = (processed_files / total_files) * 100
                    print(f"Progress: {progress:.1f}% ({processed_files}/{total_files} files)")
            except Exception as e:
                print(f"Error saat memindahkan file {img_id}: {str(e)}")
            break

print("\nStatistik hasil pemindahan file:")
print("-" * 50)
# Menampilkan statistik hasil
for category in categories:
    category_path = os.path.join(base_dest_dir, category)
    file_count = len(os.listdir(category_path))
    print(f"{category:<15} : {file_count:>5} files")

print("-" * 50)
print(f"Total file yang dipindahkan: {processed_files}")
print("\nProses selesai!")
```

```python
# Membaca file CSV
df = pd.read_csv('filtered_categories.csv')

# Definisi path
base_dir = '../Dataset'

# Definisi format penamaan untuk setiap kategori
category_prefix = {
    'Bags': 'bags',
    'Bottomwear': 'bw',
    'Dress': 'dress',
    'Headwear': 'hw',
    'Shoes': 'shoes',
    'Topwear': 'tw',
    'Watches': 'watch'
}

print("Memulai proses pengubahan nama file...")

# Proses untuk setiap kategori
for category, prefix in category_prefix.items():
    print(f"\nMemproses kategori: {category}")
    category_path = os.path.join(base_dir, category)
    
    # Mendapatkan semua file dalam folder kategori
    files = [f for f in os.listdir(category_path) if os.path.isfile(os.path.join(category_path, f))]
    
    # Mengurutkan file untuk memastikan urutan yang konsisten
    files.sort()
    
    # Mengubah nama file
    for index, filename in enumerate(files, 1):
        # Mendapatkan ekstensi file
        _, ext = os.path.splitext(filename)
        
        # Format nama baru dengan padding 4 digit
        new_name = f"{prefix}{index:04d}{ext}"
        
        # Path lengkap untuk file lama dan baru
        old_file = os.path.join(category_path, filename)
        new_file = os.path.join(category_path, new_name)
        
        try:
            os.rename(old_file, new_file)
            if index % 100 == 0:
                print(f"Progress {category}: {index}/{len(files)} files")
        except Exception as e:
            print(f"Error saat mengubah nama file {filename}: {str(e)}")

print("\nStatistik hasil pengubahan nama:")
print("-" * 50)
# Menampilkan statistik hasil
for category in category_prefix.keys():
    category_path = os.path.join(base_dir, category)
    file_count = len(os.listdir(category_path))
    print(f"{category:<15} : {file_count:>5} files")
    
    # Menampilkan contoh nama file pertama dan terakhir
    files = sorted(os.listdir(category_path))
    if files:
        print(f"{'':15} First file: {files[0]}")
        print(f"{'':15} Last file:  {files[-1]}")
    print()

print("-" * 50)
print("\nProses selesai!")

# Opsional: Menyimpan mapping ID lama ke nama file baru
mapping_data = []
for category, prefix in category_prefix.items():
    category_path = os.path.join(base_dir, category)
    files = sorted(os.listdir(category_path))
    category_df = df[df['subCategory'] == category]
    
    for index, (_, row) in enumerate(category_df.iterrows(), 1):
        mapping_data.append({
            'old_id': row['id'],
            'new_filename': f"{prefix}{index:04d}",
            'category': category
        })

mapping_df = pd.DataFrame(mapping_data)
mapping_df.to_csv('filename_mapping.csv', index=False)
print("\nMapping ID lama ke nama file baru telah disimpan di 'filename_mapping.csv'")
```

```python
import os
from PIL import Image
import pandas as pd

# Definisi ukuran yang diinginkan
TARGET_SIZE = (177, 177)

# Definisi path
base_dir = '../Dataset'
categories = ['Bags', 'Bottomwear', 'Dress', 'Headwear', 'Shoes', 'Topwear', 'Watches']

def resize_image(input_path, output_path):
    """Fungsi untuk mengubah ukuran gambar dengan mempertahankan aspek ratio"""
    try:
        with Image.open(input_path) as img:
            # Convert to RGB jika gambar dalam mode RGBA
            if img.mode == 'RGBA':
                img = img.convert('RGB')
                
            # Resize gambar
            img_resized = img.resize(TARGET_SIZE, Image.Resampling.LANCZOS)
            
            # Simpan gambar yang sudah diresize
            img_resized.save(output_path, 'JPEG', quality=95)
            return True
    except Exception as e:
        print(f"Error saat memproses {input_path}: {str(e)}")
        return False

# Statistik untuk melacak proses
total_processed = 0
total_success = 0
category_stats = {}

print("Memulai proses resize gambar...")

# Proses setiap kategori
for category in categories:
    print(f"\nMemproses kategori: {category}")
    category_path = os.path.join(base_dir, category)
    
    # Buat folder temporary untuk menyimpan hasil resize
    temp_folder = os.path.join(base_dir, f"{category}_temp")
    os.makedirs(temp_folder, exist_ok=True)
    
    # Ambil daftar file dalam folder
    files = [f for f in os.listdir(category_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    # Statistik untuk kategori ini
    category_stats[category] = {'total': len(files), 'success': 0}
    
    # Proses setiap file
    for i, filename in enumerate(files, 1):
        input_path = os.path.join(category_path, filename)
        output_path = os.path.join(temp_folder, filename)
        
        if resize_image(input_path, output_path):
            category_stats[category]['success'] += 1
            total_success += 1
        
        total_processed += 1
        
        # Tampilkan progress
        if i % 10 == 0:
            print(f"Progress {category}: {i}/{len(files)} files")

    # Ganti folder lama dengan folder baru
    import shutil
    shutil.rmtree(category_path)
    os.rename(temp_folder, category_path)

print("\nStatistik hasil resize:")
print("-" * 50)
for category, stats in category_stats.items():
    success_rate = (stats['success'] / stats['total'] * 100) if stats['total'] > 0 else 0
    print(f"{category:<15}: {stats['success']}/{stats['total']} files ({success_rate:.1f}% success)")

print("-" * 50)
print(f"Total file yang berhasil diproses: {total_success}/{total_processed}")

# Verifikasi ukuran file hasil
print("\nMemeriksa ukuran file hasil...")
verification_stats = {'correct_size': 0, 'incorrect_size': 0}

for category in categories:
    category_path = os.path.join(base_dir, category)
    files = [f for f in os.listdir(category_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    for filename in files:
        filepath = os.path.join(category_path, filename)
        with Image.open(filepath) as img:
            if img.size == TARGET_SIZE:
                verification_stats['correct_size'] += 1
            else:
                verification_stats['incorrect_size'] += 1
                print(f"Warning: {filepath} memiliki ukuran {img.size}")

print("\nHasil verifikasi ukuran:")
print(f"File dengan ukuran benar (177x177): {verification_stats['correct_size']}")
print(f"File dengan ukuran tidak sesuai: {verification_stats['incorrect_size']}")
```
